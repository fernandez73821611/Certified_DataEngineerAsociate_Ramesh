## 1. Manejo de archivos (dbutils.fs)
- **`dbutils.fs.ls(path)`** â†’ Lista archivos y carpetas en la ruta especificada.
- **`dbutils.fs.cp(source, destination, recurse=True)`** â†’ Copia archivos o carpetas.
- **`dbutils.fs.mv(source, destination, recurse=True)`** â†’ Mueve archivos o carpetas.
- **`dbutils.fs.rm(path, recurse=True)`** â†’ Elimina archivos o carpetas.
- **`dbutils.fs.head(path, maxBytes=1000)`** â†’ Muestra las primeras lÃ­neas de un archivo.
- **`dbutils.fs.put(path, contents, overwrite=True)`** â†’ Crea o sobrescribe un archivo con contenido.

---

## 2. Manejo de secretos (dbutils.secrets)
- **`dbutils.secrets.listScopes()`** â†’ Lista los *secret scopes* disponibles.
- **`dbutils.secrets.list(scope="mi_scope")`** â†’ Lista las claves dentro de un *scope*.
- **`dbutils.secrets.get(scope="mi_scope", key="mi_clave")`** â†’ Obtiene el valor de un secreto.

---

## 3. EjecuciÃ³n de notebooks (dbutils.notebook)
- **`dbutils.notebook.run("notebook_hijo", timeout_seconds=60, arguments={"param":"valor"})`**  
  Ejecuta otro notebook como proceso independiente, pasando parÃ¡metros y devolviendo un valor de salida.

---

## 4. Widgets para parametrizaciÃ³n (dbutils.widgets)
- **`dbutils.widgets.text("param1", "valor_defecto")`** â†’ Crea un widget de texto.
- **`dbutils.widgets.dropdown("param2", "opcion1", ["opcion1","opcion2"])`** â†’ Crea un menÃº desplegable.
- **`dbutils.widgets.get("param1")`** â†’ Obtiene el valor ingresado en el widget.

---

## 5. Jobs y tareas (dbutils.jobs)
- **`dbutils.jobs.taskValues.set(key="resultado", value="ok")`** â†’ Guarda un valor en el contexto del job.
- **`dbutils.jobs.taskValues.get(taskKey="mi_tarea", key="resultado", debugValue="default")`** â†’ Recupera un valor de otra tarea.

---

## 6. Credenciales (dbutils.credentials)
- **`dbutils.credentials.showCurrentRole()`** â†’ Muestra el rol actual con el que se ejecuta el cluster.
- **`dbutils.credentials.showRoles()`** â†’ Lista los roles disponibles.

---

# âœ… Buenas prÃ¡cticas en producciÃ³n
- Usar **`dbutils.secrets`** para credenciales en lugar de hardcodearlas.
- Modularizar notebooks con **`dbutils.notebook.run()`** en lugar de `%run` cuando se requiera aislamiento.
- Parametrizar notebooks con **`dbutils.widgets`** para mayor flexibilidad.
- Manejar archivos en **DBFS o montajes de ADLS/S3** con `dbutils.fs`.
