# Depuración de Notebooks en Databricks

Databricks ofrece la posibilidad de habilitar un **depurador interactivo** para notebooks en Python, lo que facilita encontrar y corregir errores en tu código.

---

## Cómo habilitar el depurador
1. Ve a **Settings → Developer**.
2. Activa la opción **Enable Python notebook interactive debugger**.

---

## Requisitos
- **Runtime soportado:** 13.3 LTS o superior.  
- **Lenguajes soportados:** Actualmente solo **Python**.

---

## Funcionalidades principales
El depurador interactivo permite:

1. **Breakpoints (puntos de interrupción):**  
   Detener la ejecución en una línea específica para inspeccionar el estado del programa.

2. **Ejecución paso a paso:**  
   Avanzar línea por línea para entender el flujo del código.

3. **Explorador de variables:**  
   Visualizar los valores actuales de las variables en el entorno de ejecución.

4. **Consola de depuración:**  
   Ejecutar comandos y expresiones mientras el programa está detenido, para probar soluciones rápidas.

---

## Beneficios
- Facilita la identificación de errores lógicos y de ejecución.  
- Permite comprender mejor el comportamiento del código.  
- Acelera el proceso de corrección y mejora la calidad de los notebooks.


```python
# Ejemplo de depuración en Databricks Notebook

def calcular_promedio(numeros):
    total = sum(numeros)        # Aquí puedes poner un breakpoint
    cantidad = len(numeros)     # Otro breakpoint útil
    promedio = total / cantidad
    return promedio
```

```python

# Lista de números de prueba
valores = [10, 20, 30, 40, 50]

# Llamada a la función
resultado = calcular_promedio(valores)

print("El promedio es:", resultado)
```
