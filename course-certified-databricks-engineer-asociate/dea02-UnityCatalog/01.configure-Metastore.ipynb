## CREAR UN CATALOGO DATOS 


```sql
-- PRIMERAMTNE ENUMERAMOS TODOS LOS CATALOGOS DENTRO DE NUESTA UNIDAD DE METASTORE
SHOW CATALOGS
```

```sql
-- ESTO ES PARA LISTAR LOS SCHEMAS CREADOS EN LA UNIDAD DE MESTORE
SHOW SCHEMAS
```

```sql
-- ESTO NOS PERMITE CREA UN CATALOGO DENTRO DE METASTORE
CREATE CATALOG  IF NOT EXISTS gizmobox
      MANAGED LOCATION 'abfss://gizmobox@innovatecflashadlsg2.dfs.core.windows.net'
      COMMENT "Creandog mi catalogo  gizmobox"
 
```

```sql
-- PARA VER EL LISTADO DE CATALOGOS CREADOS
SHOW CATALOGS;
```

```sql
-- PARA VER EN QUE CATALOGO ESTAMOS POSICIONADOS
SELECT current_catalog();
```

```sql
-- cambiamos de catalogo
USE CATALOG gizmobox;
```

```sql
--CREAMOS UN SCHEMA PARA CADA  CAPA EN LA ARQUITECTURA MEDALLON
-- COM BRONZE, SILVER, GOLD
USE CATALOG gizmobox;
CREATE SCHEMA IF NOT EXISTS landing
    COMMENT 'Creacion del schema lanidng dentro del lakehouse'
    MANAGED LOCATION 'abfss://gizmobox@innovatecflashadlsg2.dfs.core.windows.net/landing/'
```

## Ahora creamos un volumen como capa de abstraccion en landig

```sql
USE CATALOG gizmobox;
USE SCHEMA landing;
CREATE EXTERNAL  VOLUME IF NOT EXISTS operational_data
    LOCATION 'abfss://gizmobox@innovatecflashadlsg2.dfs.core.windows.net/landing/operational_data/'
    COMMENT 'Creando  mi capa de abstraccion para mi datos'
```

```sql
%fs ls /Volumes/gizmobox/landing/operational_data
```
