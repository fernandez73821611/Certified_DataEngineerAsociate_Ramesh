```python
# Para poder utlizar pyspark y ejecutra comando de sql lo hacemos de la siguiente manera
df_customers=spark.sql("SELECT *FROM JSON.`/Volumes/gizmobox/landing/operational_data/customers/customers_2024_10.json`")
```

```python
display(df_customers)
```

```python
#Tambien podemos crear vistas con pyspark

df=spark.sql('''
             CREATE OR REPLACE TEMP VIEW v_customers
             AS
             SELECT *FROM JSON.`/Volumes/gizmobox/landing/operational_data/customers/customers_2024_10.json`
             ''')

```

```python
%sql
SELECT *FROM v_customers;
```
