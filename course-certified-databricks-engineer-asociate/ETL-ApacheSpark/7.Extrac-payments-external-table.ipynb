# Estraer datos de payments mediante una tabla externa

En una tabla externa tu puedes especificar columnas tipos de datos,delimitador,header , entre otros, las tablas exteranas se crean ensima de tus datos y  todo lo relacionado con la tabla externa como nombre, columnas , tipo de dato se guardan en metadatos en unity catalog, por la cual al intenta eliminar una tabla externa solo se eliminan esos metadatos mas no los datos, la diferencia con tablas managemed

Otra apreciasion es que si llegan registros nuevos o archivos , tienes que actualizar esos metadatos o la tabla externa par aque pueda recien utilizar esos datos , o si eliminas un archivo aparecera error

```sql
%fs ls "abfss://gizmobox@innovatecflashadlsg2.dfs.core.windows.net/landing/external_data/payments"
```

LA LOCATION o ubicacion es lo que le hace ser una tabla externa , porque si no pones la location se convierte en una tabla managemend

```sql
CREATE TABLE IF NOT EXISTS gizmobox.bronze.payments
(
  payment_id INTEGER,
  order_id INTEGER,
  payment_timestamp TIMESTAMP,
  payment_status INTEGER,
  payment_method STRING
  
)
USING CSV
options (
  header="true",
  delimiter=","
)
LOCATION "abfss://gizmobox@innovatecflashadlsg2.dfs.core.windows.net/landing/external_data/payments"
```

```sql
SELECT *FROM gizmobox.bronze.payments;
```

```sql
-- PARA NOSOSTROS SABER SI ESTA ES UNA TABLA EXTERNA O ADMINISTRADA 
DESCRIBE EXTENDED gizmobox.bronze.payments;
```

### si tu eliminar al algun archivo te aparece error , tienes que actualizar la tabla

### igual si agregas registros o archivos necitas actualizar la tabla para ver esos cambios

```sql
-- PARA PODER ACTUALIZAR LOS METADASTOS EN EL METASTORE
REFRESH TABLE gizmobox.bronze.payments;
-- LO QUE HACE ES REFRESCAR LO METDATASTOS DE LA TABLA EN EL METASTORE 
```

```sql
-- AL ELIMINAR LA TABLA SOLO SE ELIMINA LOS METADATOS ALMACENADOS EN EL STORAGE DEL METASTORE  MAS NO LOS DATOS --FISICOS DE LA TABLA
DROP TABLE IF EXISTS gizmobox.bronze.payments
```
