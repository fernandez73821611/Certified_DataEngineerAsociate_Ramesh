### TOTAL DE ORDERS

```sql
SELECT *FROM gizmobox.silver.orders limit 10;
```

### Total de articulos comprados

```sql
SELECT 
date_format(transaction_timestamp,"yyyy-MM") as order_mount,
customer_id,
count(distinct(order_id)) as total_orders,
sum(quantity) as total_items_bought,
sum(price*quantity)as total_amount
FROM gizmobox.silver.orders
group by order_mount,customer_id;
```

### Importe total gastado

```sql
CREATE TABLE IF NOT EXISTS gizmobox.gold.order_sumary_monthly

AS
SELECT 
date_format(transaction_timestamp,"yyyy-MM") as order_mount,
customer_id,
count(distinct(order_id)) as total_orders,
sum(quantity) as total_items_bought,
sum(price*quantity)as total_amount
FROM gizmobox.silver.orders
group by order_mount,customer_id;
```
