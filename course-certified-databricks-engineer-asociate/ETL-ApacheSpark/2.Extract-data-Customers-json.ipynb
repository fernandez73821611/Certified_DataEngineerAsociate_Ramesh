### Query folder json

```sql
-- SI SOLO ESPECIFICAS EL FOLDER O CARPETA LEERA TODOS LOS ARCHIVOS
SELECT *FROM JSON.`/Volumes/gizmobox/landing/operational_data/customers`
```

```sql
-- la que esta en desuso pero es bueno menccionar por si venga en el examen: esto esta en desuso o anticuado
SELECT input_file_name() as file_path,* FROM JSON.`/Volumes/gizmobox/landing/operational_data/customers`
```

```sql
-- AL MOMENTO DE REALIZAR UN EXTRACT DENTRO DE UN PROYECTO ETL ES NECEARIO AGREGAR FUNCIONES DE AUDITORIA,
-- PARA RASTREAR LOS ARCHIVOS DE DONDE PROCEDEN EN CASO MAS ADELANTE EL PROCESAMIENTO FALLE

SELECT _metadata.file_path,* FROM JSON.`/Volumes/gizmobox/landing/operational_data/customers`
```

```sql
-- si tu quieres tener varios metadatos de tu archivo utilizamos _metadata
-- aca podemos ver el file_path, file_name, la modificacion del archivo es muy genial
SELECT _metadata,* FROM JSON.`/Volumes/gizmobox/landing/operational_data/customers`
```
