# EXTRACT DATA ORDERS JSON COMPLEX- CON PROBLEMAS DE CALIDAD


Debido a que este archivo tiene problemas de calidad de datos  se debe reprocesar antes de cargar a un fomato json, por la cual lo que se sujiere es que se lea como un txt,
cuando  tienen problemas de calidad tienes que tener una logica de analisis personalizado antes de hacer el extrar , pero en este caso sera cargado como un txt para poder tener todos los datos y luego aplicar la logica


### Primeramente utilizaremos el formato JSON para extraer los datos

```sql
SELECT *FROM JSON.`/Volumes/gizmobox/landing/operational_data/orders/orders*.json`
```

Lo pasamos en un formato txt para luego reprocesarlo y darle un formato con la analitica JSON

```sql
SELECT *FROM TEXT.`/Volumes/gizmobox/landing/operational_data/orders/orders*.json`
```

```sql
CREATE OR REPLACE VIEW gizmobox.bronze.v_orders
AS
SELECT *FROM TEXT.`/Volumes/gizmobox/landing/operational_data/orders/orders*.json`
```

```sql
SELECT *FROM gizmobox.bronze.v_orders;
```
