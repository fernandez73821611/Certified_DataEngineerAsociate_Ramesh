# UDF(FUNCIONES DEFINIDAS POR  USUARIOS)

LAS FUNCIONES DATABRICKS RECOMIENDA MEJOR ESCRIBIRLOS EN SQL QUE EN PYTHON

### Sintaxis
CREATE OR REPLACE FUNCTION  catalogo.schema.udf_name(paramtros datatype)  
return type  
return expression

```sql
CREATE OR REPLACE FUNCTION gizmobox.default.udf_nombre_completo(nombre STRING, apellido STRING)
RETURNS STRING
RETURN CONCAT(initcap(nombre), ' ',initcap(apellido))
```

```sql
SELECT gizmobox.default.udf_nombre_completo("Joseph","Fernandez")
```

```sql
-- SI yo deceo ver propiedas de la  fucion por ejemplo ne que storage location se a guardado
DESC FUNCTION EXTENDED gizmobox.default.udf_nombre_completo;
```
