### Debido a que en la culumna de items hay arrays duplicados utilizaremos la siguiente funcion: array_distinct(arreglo)


```sql
SELECT 
  json_value.order_id,
  json_value.order_status,
  json_value.payment_method,
  json_value.total_amount,
  json_value.transaction_timestamp,
  json_value.customer_id,
  array_distinct(json_value.items) as items
FROM gizmobox.silver.orders_json
limit 10;
```

### Como vemos en la columna items siguen abiendo dentro de un arreglo arreglos anidados por lo cual par atrabajar esto se utiliza la funcion explode(collection)

Lo que hace explod ees dibir al celda de items que tengas mas de uno en otra fila

```sql
SELECT 
  json_value.order_id,
  json_value.order_status,
  json_value.payment_method,
  json_value.total_amount,
  json_value.transaction_timestamp,
  json_value.customer_id,
  explode(array_distinct(json_value.items)) as item
FROM gizmobox.silver.orders_json
```

```sql
CREATE OR REPLACE TEMPORARY VIEW tv_orders_exploded

AS 

SELECT 
  json_value.order_id,
  json_value.order_status,
  json_value.payment_method,
  json_value.total_amount,
  json_value.transaction_timestamp,
  json_value.customer_id,
  explode(array_distinct(json_value.items)) as item
FROM gizmobox.silver.orders_json
```

```sql
SELECT 
order_id,
order_status,
payment_method,
total_amount,
transaction_timestamp,
customer_id,
item.item_id,
item.name,
item.price,
item.quantity,
item.category,
item.details.brand,
item.details.color

FROM tv_orders_exploded;
```

```sql
CREATE TABLE IF NOT EXISTS gizmobox.silver.orders
AS

SELECT 
order_id,
order_status,
payment_method,
total_amount,
transaction_timestamp,
customer_id,
item.item_id,
item.name,
item.price,
item.quantity,
item.category,
item.details.brand,
item.details.color

FROM tv_orders_exploded;
```
