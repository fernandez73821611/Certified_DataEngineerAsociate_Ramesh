# Creacion de vistas o views en la capa bronze

Esta es conocida como una vista permente o estandar es decir siempre va a estar disponible incluso el cluster se reinicie

```sql
CREATE OR REPLACE VIEW gizmobox.bronze.v_customers 

AS SELECT *,_metadata.file_path as file_path FROM JSON.`/Volumes/gizmobox/landing/operational_data/customers`
```

```sql
SELECT *FROM gizmobox.bronze.v_customers;
```

# CREATE TIPO DE VISTA TEMPOTAL- TEMPORARY VIEW

La View Temp o vista temporal solo exite en el inicio de sesion o sparksession quiere decir que si tu noteboock de desvincular del cluster o cierras session de tu worspace se eliminara

```sql
CREATE OR REPLACE TEMPORARY VIEW tv_customers 

AS SELECT *,_metadata.file_path as file_path FROM JSON.`/Volumes/gizmobox/landing/operational_data/customers`
```

```sql
SELECT *FROM tv_customers;
```

# CREATE TIPO DE VISTA GLOVAL- GLOBAL TEMPORARY VIEW

Este tipop de vista temporal exite a nivel de cluster y puede ser utilizado por diferentes notebooks conectados a este cluster , se elimina cunado el el cluster se apaga o se reinicia
-lo mas importante esque estas vistas temporales nos son objetos de Unity Catalog

```sql
CREATE OR REPLACE TEMP VIEW gtv_customers 

AS SELECT *,_metadata.file_path as file_path FROM JSON.`/Volumes/gizmobox/landing/operational_data/customers`
```

```sql
-- PRARA PODER NOSOTROS ACCEDER A UNA VISTA TEMPOPRAL GLOBA , NECESITAMOS ACCEDER DESD UN ESQUEMA DEL SISTEMA 
-- QUE YA EXISTE 
SELECT *FROM global_temp.gtv_customers;
```
