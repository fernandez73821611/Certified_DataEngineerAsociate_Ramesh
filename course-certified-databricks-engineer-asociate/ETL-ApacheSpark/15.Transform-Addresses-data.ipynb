# TRANSFORM ADDRESSES DATA 

### Debido a que cada cliente tienes dos direcciones , lo que vamos a hacer es convinar estas columnas

## PIVOTE :Convierte filas en columnas 

CLAVE EXAMEN: SIEMPRE LA FUNCION PIVOT TIENES QUE PONERLO UNA AGREGACION COMO COUNT, MIN ,MAX , AVG, etc no olvidar y luego sigue el for

```sql
SELECT 
customer_id,
address_type,
address_line_1,
city,
state,
postcode
from gizmobox.bronze.v_addresses;
```

```sql
SELECT *FROM( SELECT customer_id,
                address_type,
                address_line_1,
                city,
                state,
                postcode
            from gizmobox.bronze.v_addresses)
PIVOT(
MAX(address_line_1) as address_line_1,
MAX(city) as city,
MAX(state) as state,
MAX(postcode) as postcode
FOR address_type IN ('shipping','billing')
);
```

```sql
CREATE OR REPLACE TABLE gizmobox.silver.adreesses AS
SELECT *FROM( SELECT customer_id,
                address_type,
                address_line_1,
                city,
                state,
                postcode
            from gizmobox.bronze.v_addresses)
PIVOT(
MAX(address_line_1) as address_line_1,
MAX(city) as city,
MAX(state) as state,
MAX(postcode) as postcode
FOR address_type IN ('shipping','billing')
);
```

```sql
SELECT *FROM gizmobox.silver.adreesses;
```
