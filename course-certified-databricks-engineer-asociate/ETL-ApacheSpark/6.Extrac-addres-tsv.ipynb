# Estraemos el archivo Address que es tipo Tsv=> read_files()

Ahora para poder leer este tipo de archivo como sabemos este tiene una cabecera y tambien esta separador por un delimitador que es el tap , por la cual tenemos que utilizar una funcion de valot de tabla read_files para poder pasar esos parametros


```sql
SELECT * FROM
read_files('/Volumes/gizmobox/landing/operational_data/addresses',
         format=>'csv',
         delimiter=>"\t",
         header => TRUE
        )
```

```sql
CREATE OR REPLACE VIEW gizmobox.bronze.v_addresses
AS
SELECT * FROM
read_files('/Volumes/gizmobox/landing/operational_data/addresses',
         format=>'csv',
         delimiter=>"\t",
         header => TRUE
        )
```

```sql
SELECT *FROM gizmobox.bronze.v_addresses;
```
