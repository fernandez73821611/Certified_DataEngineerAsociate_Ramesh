# DATA PROFILING IN DATABRICKS

### “Explorar los datos antes de transformarlos o usarlos.”
Un Data Engineer hace profiling para detectar problemas como:

valores nulos

duplicados

tipos incorrectos

rangos inválidos

outliers

- Nulls inesperados	               isNull()
- Tipos incorrectos	             printSchema()
- Valores fuera de rango	         describe()
- Duplicados	                  dropDuplicates()
- Categorías raras	           distinct() / groupBy()

## Perfile data manually

Contamos los clientes o customers por el id

```python
%sql
SELECT COUNT(*) AS NUM_REGISTROS ,COUNT(customer_id) as Num_Clientes,COUNT(email)as Num_Email FROM gizmobox.bronze.v_customers;
```

Para ver si no hay duplicados en los id customers ]

```python
%sql
SELECT count_if(customer_id is null) as `Customer_id_distinct` FROM gizmobox.bronze.v_customers
```

```python
%sql
SELECT count(*) as `Customer_id_distinct` FROM gizmobox.bronze.v_customers where customer_id is null
```

PARA VER SI HAY DATOS DUPLICADOS

```python
%sql
SELECT COUNT(*) AS total_num_register, COUNT(DISTINCT customer_id) AS distintos_customers
FROM gizmobox.bronze.v_customers where customer_id is not null;
```
