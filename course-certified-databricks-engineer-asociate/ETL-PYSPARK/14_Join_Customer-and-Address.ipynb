# Join o union de la tabla customer y Address

```python
df_address=spark.table("gizmobox.silver.py_address")
```

```python
df_customers=spark.table("gizmobox.silver.py_customers")
```

```python
display(df_address.limit(5))
```

```python
display(df_address.limit(5))
```

```python
df_customer_address=(
    df_customers.join(df_address,'customer_id',how='inner')
    .select(
        "customer_id",
        "customer_name",
        "date_of_birth",
        "email",
        "member_since",
        "telephone",
        "billing_address_line_1",
        "billing_city",
        "billing_state",
        "billing_postcode",
        "shipping_address_line_1",
        "shipping_city",
        "shipping_state",
        "shipping_postcode"
    )
)
display(df_customer_address.limit(5))
```

```python
df_customer_address.writeTo("gizmobox.gold.py_customer_address").createOrReplace()
```
