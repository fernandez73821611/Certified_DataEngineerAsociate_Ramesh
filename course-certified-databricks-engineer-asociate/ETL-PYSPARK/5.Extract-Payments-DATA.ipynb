# Extracion de datos de lata hive_metastore de Payments<>

```python
schema_pay='payment_id INTEGER,order_id INTEGER'
```

```python
#Debido a que no cuenta con una cabecera ahora lo agregamos
from pyspark.sql.types import * 
shema_payments=StructType([
    StructField("payment_id",IntegerType()),
    StructField("order_id",IntegerType())
])
```

```python
df=(
    spark.read.format("csv")
            .option("delimiter",",")
            .option("header","true")
            .load("ruta_de_la_carpeta")
            .schema(schema_pay)
)
```

```python
df.writeTo("gizmobox.bronze.py_payments").createOrRepalace()
```
