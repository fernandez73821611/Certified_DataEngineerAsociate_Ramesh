# Transformacion de la tabla Address

```python
df_address=spark.read.table("gizmobox.bronze.py_address")
```

Como sabemos los datos de cliente tienen dos direcciones y queremos tenerlo en un solo registro ,Lo que aremos es tomar una columana pivote que es addresses par luego pivotarlo en columnas

### GroypeData.pivot(pivot_col,values=None): En pyspark se utiliza esta funcionn a diferencia de SQL

```python
from pyspark.sql.functions import max
df_1=(
df_address
.groupBy('customer_id')
.pivot('address_type',['billing','shipping'])
.agg(
  max('address_line_1').alias('address_line_1'),
  max('city').alias('city'),
  max('state').alias('state'),
  max('postcode').alias('postcode')
)

)

display(df_1)
```

```python
df_1.writeTo("gizmobox.silver.py_address").createOrReplace()
```
