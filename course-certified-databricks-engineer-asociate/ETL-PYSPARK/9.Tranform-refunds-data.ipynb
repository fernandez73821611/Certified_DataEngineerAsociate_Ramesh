# Transformacion de la tabla Refunds Data que consiste en dividir una columna de string

```python
df_refunds=spark.read.table("gizmobox.bronze.refunds")
```

```python
df_refunds_1=df_refunds.select(
    'refund_id',
    'payment_id',
    date_format('refund_timestamp','yyyy-MM-dd').alias('refund_date'),
    date_format('refund_timestamp','HH:mm:ss'),
    'refund_amount',
    split('refund_reason',':')[0].alias("refund_reason"),
    split('refund_reason',':')[1].alias("refund_source")
    )
```

```python
df_refunds_1.writeTo("gizmobox.silver.py_refunds").createOrReplace()
```
