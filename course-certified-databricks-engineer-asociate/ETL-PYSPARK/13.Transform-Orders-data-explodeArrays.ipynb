# Vamos a transformar el objeto json de la tabla orders

```python
df_orderss_json=spark.table("gizmobox.silver.py_orders_json")
```

```python
display(df_orderss_json.limit(5))
```

PARA ACCEDERA A CADA ELEMENTO

```python
from pyspark.sql import functions as F
```

```python
df_orders_1=(
  df_orderss_json.select(
    "json_value.order_id",
    "json_value.order_status",
    "json_value.payment_method",
    "json_value.total_amount",
    "json_value.transaction_timestamp",
    "json_value.customer_id",
    "json_value.items"
  )
)
```

```python
display(df_orders_1.limit(5))
```

```python
df_orders_1=(
  df_orderss_json.select(
    "json_value.order_id",
    "json_value.order_status",
    "json_value.payment_method",
    "json_value.total_amount",
    "json_value.transaction_timestamp",
    "json_value.customer_id",
    F.array_distinct("json_value.items").alias("items")
  )
)
```

```python
display(df_orders_1.limit(5))
```

```python
df_orders_exploed=(
    df_orders_1.select(
        "order_id",
        "order_status",
        "payment_method",
        "total_amount",
        "transaction_timestamp",
        "customer_id",
        F.explode("items").alias("item")
    )
)
```

```python
display(df_orders_exploed.limit(5))
```

```python
df_f=(
    df_orders_exploed.select(
        "order_id",
        "order_status",
        "payment_method",
        "total_amount",
        "transaction_timestamp",
        "customer_id",
        "item.item_id",
        "item.name",
        "item.price",
        "item.quantity",
        "item.category",
        "item.details.brand",
        "item.details.color"
        
)
)

display(df_f.limit(5))
```

```python
df_f.writeTo("gizmobox.silver.py_orders").createOrReplace()
```
