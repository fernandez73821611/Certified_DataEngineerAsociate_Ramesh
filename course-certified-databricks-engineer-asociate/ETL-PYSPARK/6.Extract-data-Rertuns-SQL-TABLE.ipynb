# Extraccion de datos de la tabla de AZURE SQL DATABSE Refunds mediante JDBC

```python
df=(
  spark.read.format("jdbc")
  .option("url", "jdbc:postgresql://database.databricks.training:5432/training")
  .option("dbtable", "customers")
  .option("user", "training")
  .option("password", "training")
  .load()
)
display(df)
```

```python
df.writeTo("gizmobox.bronzw.py_refunds").createOrReplace()
```
