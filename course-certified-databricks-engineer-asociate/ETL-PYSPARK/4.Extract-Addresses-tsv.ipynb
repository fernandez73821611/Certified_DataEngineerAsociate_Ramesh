# Etraccion de datos del archivo Address que es un formato Tsv VALORES SEPARADOS POR COMA

```python
df_address=spark.read.format("csv").load("/Volumes/gizmobox/landing/operational_data/addresses/*.tsv")
```

```python
display(df_address.limit(10))
```

```python
# Para que pueda leer tines que si o si decirlo a spark que cuenta con un separador y un header
df_address_result=spark.read.format("csv").option("sep","\t").option("header","true").load("/Volumes/gizmobox/landing/operational_data/addresses/*.tsv")
```

```python
display(df_address_result.limit(10))
```

```python
df_address_result.writeTo("gizmobox.bronze.py_address").createOrReplace()
```

```python
display(
    spark.table("gizmobox.bronze.py_address").limit(10)
)
```
