# Transformacion de la tabla pagos Payments data

```python
df_payments=spark.read.table("gizmobox.bronze.payments")
```

```python
display(df_payments.limit(10))
```

EXTRAENDO FECHA Y HORA DE PAGO

```python
from pyspark.sql import functions as F

df_payments=df_payments.select(
     'payment_id',
     'order_id',
      F.date_format('payment_timestamp','yyyy-MM-dd').cast("date").alias("payment_date"),
      F.date_format('payment_timestamp','HH:mm:ss').alias("payment_time"),
     'payment_status',
     'payment_method'
 )
display(df_payments)
```

```python
df_payments=(
    
    df_payments.select(
        'payment_id',
        'order_id',
         'payment_date',
         'payment_time',
         F.when(df_payments.payment_status==1,'Success')
         .when(df_payments.payment_status==2,'Pending')
          .when(df_payments.payment_status==3,'Cancelled')
          .when(df_payments.payment_status==4,'Failed')
          .alias('payment_status'),
        'payment_method'
    )
)

```

```python
display(df_payments.limit(10))
```

```python
df_payments.writeTo("gizmobox.silver.py_payment").createOrReplace()
```
