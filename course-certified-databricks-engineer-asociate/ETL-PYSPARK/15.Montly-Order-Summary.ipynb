# Montly Orders Summary

1. Total orders
2. Total items bought
3. Total amount spent

```python
df_orders=spark.read.table("gizmobox.silver.py_orders")
```

```python
display(df_orders.limit(10))
```

```python
from pyspark.sql import functions as F
df_summary=(
    df_orders
        .withColumn("order_mount",F.date_format("transaction_timestamp", "yyyy-MM"))
        .groupBy("order_id","customer_id")
        .agg(
         F.countDistinct("order_id").alias("total_orders"),
        F.sum("quantity").alias("total_items_bought"),
        F.sum(F.col("price") * F.col("quantity")).alias("total_mount"),
  
        )
)
```

```python
display(df_summary.limit(10))
```

```python
df_summary.writeTo("gizmobox.gold.py_summary").createOrReplace()
```
